```python
#Applying PCA to the data

# Standardize the data
X = StandardScaler().fit_transform(arrestsdf.values)
```

```python
# Apply PCA with 2 components
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
```

```python
# Create a new dataframe for the principal components
columns = ['PC1', 'PC2']
principalDf = pd.DataFrame(data=principalComponents, columns=columns)
```

```python
# Merge the principal components with the original dataframe
finalDf = pd.concat([principalDf, arrestsdf[['Murder', 'Assault', 'UrbanPop', 'Rape']]], axis=1)

#finalDf.head()
```

```python
def biplot(score, coeff, labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())

    fig, ax = plt.subplots()    

    ax.scatter(xs * scalex,ys * scaley,s=5)

    for i in range(0,len(xs)):
        txt = arrestsdf.index[i] # City names
        ax.annotate(txt, (xs[i]* scalex, ys[i]* scaley))

    for i in range(n):
        ax.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            ax.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'green', ha = 'center', va = 'center')
        else:
            ax.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
 
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid()

biplot(principalComponents, np.transpose(pca.components_[0:2, :]), labels=['Murder', 'Assault', 'UrbanPop', 'Rape'])
```

```python
# Standardize the data
X = StandardScaler().fit_transform(arrestsdf.iloc[:, 1:].values)

# Perform PCA
pca = PCA()
principalComponents = pca.fit_transform(X)

# Create a dataframe with the principal components and city names
df_pca = pd.DataFrame(data = principalComponents[:,0:2], columns = ['PC1', 'PC2'])
df_pca['City'] = arrestsdf.iloc[:, 0].values
```

```python
# Plot the biplot
biplot(principalComponents[:,0:2],np.transpose(pca.components_[0:2, :]),labels=arrestsdf.columns.values[1:])
plt.show()
```

```python
# from a high-level we can see that urban population has a negative correlation with crime
# indicating that areas with lower urban population may lead to lower crime levels

# this is not necessarily true accross the board - just a high level observation
# cities can start to be grouped according to how safe they are (where crime levels are lower)
# as well as where urban populations are lower
```

```python
# Fit PCA on standardized data
pca = PCA().fit(X)

# Get variance explained by each principal component
var = pca.explained_variance_ratio_

# Get cumulative sum of variances
cum_var = np.cumsum(var)

# Plot cumulative variance explained
plt.plot(cum_var)
plt.title('Cumulative Variance Explained by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.show()
```

```python
#plotting a scree plot
plt.plot(pca.explained_variance_ratio_, 'bo-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.show()
```

```python
# Perform hierarchical clustering
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model = model.fit(df_pca)

# Plot dendrograms
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.suptitle('Hierarchical Clustering Dendrogram')

ax1.set_title("Single")
dendrogram(linkage(df_pca, method='single'), labels=df_pca.index, ax=ax1) 

ax2.set_title("Average")
dendrogram(linkage(df_pca, method='average'), labels=df_pca.index, ax=ax2)

ax3.set_title("Complete")
dendrogram(linkage(df_pca, method='complete'), labels=df_pca.index, ax=ax3)

plt.show()
```

```python
# Plot dendrogram for average linkage 
plt.title("Average linkage")
dendrogram(linkage(df_pca, method='average'), labels=df_pca.index)
plt.xticks(rotation=90)
plt.yticks(fontsize=8)
plt.show()
```

```python
# the elbow method is used to determine the number of clusters to group the data in
# where the 'elbow' bends below aligns to the spread of the dendrogram above 

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_pca.iloc[:, 0:2])
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Perform k-means clustering with k=4
# Fit k-means
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(X)

# Add cluster labels to the dataframe
df_pca['Cluster'] = labels

# Extract the first two principal components
x = principalComponents[:,0]
y = principalComponents[:,1]

# Plot clusters
fig, ax = plt.subplots()
colours = 'rbgy'
for i in range(0,k):
    ax.scatter(x[labels == i], y[labels == i], c = colours[i]) 

for i in range(0,len(x)):
    txt = arrestsdf.index[i] # City names
    ax.annotate(txt, (x[i], y[i]))

ax.set_title("K-Means cluster plot")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.legend()
```
